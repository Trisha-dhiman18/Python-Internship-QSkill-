# Student Performance Analysis using Pandas & Matplotlib

## Internship Task 1

### Objective
The objective of this project is to analyze student performance data using Python libraries such as Pandas and Matplotlib. The analysis focuses on understanding score distributions, relationships between subjects, and the impact of demographic and educational factors on student performance.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Dataset Loading
The dataset contains student demographic details and their performance in mathematics, reading, and writing.


```python
df = pd.read_csv("study_performance.csv")
```

## Statistical Analysis
Average scores were calculated to understand overall student performance.


```python
df[['math_score', 'reading_score', 'writing_score']].mean()
```

## Gender-wise Performance Analysis
This section analyzes the average scores of students based on gender.


```python
gender_scores = df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean()

gender_scores.plot(kind='bar')
plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```

## Impact of Test Preparation Course
The following visualization shows how test preparation affects student scores.


```python
prep_scores = df.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score']].mean()

prep_scores.plot(kind='bar')
plt.title("Impact of Test Preparation Course on Scores")
plt.xlabel("Test Preparation Course")
plt.ylabel("Average Score")
plt.show()
```

## Relationship Between Math and Reading Scores


```python
plt.scatter(df['math_score'], df['reading_score'])
plt.title("Math Score vs Reading Score")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

## Correlation Analysis
A heatmap is used to visualize correlations between numerical features.


```python
correlation = df[['math_score', 'reading_score', 'writing_score']].corr()

plt.imshow(correlation)
plt.colorbar()
plt.xticks(range(3), correlation.columns)
plt.yticks(range(3), correlation.columns)
plt.title("Correlation Heatmap of Scores")
plt.show()
```

## Test Preparation Course Distribution


```python
df['test_preparation_course'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Test Preparation Course Participation")
plt.ylabel("")
plt.show()
```

```python
df['lunch'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Lunch Type Distribution")
plt.ylabel("")
plt.show()
```

## Key Insights & Observations

- Students perform better on average in reading and writing than in mathematics.
- Test preparation courses have a positive impact on student performance across all subjects.
- A strong positive correlation exists between reading and writing scores.
- Math and reading scores also show a positive relationship, indicating consistent academic performance.
- Gender-based differences in subject-wise performance are observable.
- The dataset contains a nearly balanced gender distribution.
- A significant portion of students did not complete the test preparation course.
- Most students receive a standard lunch, which may influence academic performance.

This analysis demonstrates how basic data analysis and visualization techniques can be applied to extract meaningful insights from educational data. The findings can help educators and institutions understand performance patterns and improve academic outcomes.
